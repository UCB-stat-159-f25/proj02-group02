### Exploratory Data Analysis

Replicate the plots below using the hints specified. For each plot, provide some commentary describing the results/anything interesting you might see.

#### Number of Speeches per President

```python
# Hint - use value_counts() on the President column
# Hint - sort in order of dataframe
```

```python
# Plot 
# Hint - use the .plot() method for Pandas Series, make sure all presidents show up on x-axis
order = sou.groupby('President')['Year'].max().sort_values(ascending=False).index
counts = sou['President'].value_counts().reindex(order)
plt.figure(figsize=(15,5))
counts.plot(kind='bar')
plt.title('Number of Speeches per President')
plt.xlabel('President')
plt.ylabel('Count')
plt.xticks(rotation=90)
```

Commentary) 

Modern presidents usually give one speech every year, so recent presidents all have similar counts.

Older presidents, especially before the Civil War, often gave fewer speeches because the State of the Union was not yet a regular tradition.


A few presidents like FDR, have very high counts because they served much longer than a normal two-term presidency.

#### Number of Speeches per Year

```python
# Hint - Use value counts and sort by years
sou['Year'].value_counts().sort_index().plot()
plt.title('Number of State of the Union Speeches per Year')
plt.xlabel('Year')
plt.ylabel('Count')
```

commentary)

In the very early years of the country there are some blips of years that had more than one State of the Union message, which explains the higher counts around the late 1700s.


For most of U.S. history, each year only had one speech, which is why the line stays flat at 1 across almost the entire timeline.


The small bumps around the 1950s–1970s come from presidents occasionally delivering extra messages or special addresses in those years.

#### Word Count Distribution 

```python
# Hint - try seaborn.histplot()
sns.histplot(sou['Word Count'])
plt.title('Distribution of State of the Union Speech Word Counts')
plt.xlabel('Word Count')
plt.ylabel('Count')
```

Commentary)

Most speeches are around the range of 3,000 and 10,000 words. This is probably the “normal” length considered for a State of the Union.


There are a couple speeches that are much longer reaching over 20,000 or even 30,000 words, which stand out as rare outliers.


The distribution is skewed to the right, meaning long speeches are uncommon but do happen occasionally. Most speeches are shorter. 


#### Word Count Distribution over Year

```python
# Hint: try seaborn.rugplot()
sns.scatterplot(x='Word Count', y='Year', data=sou)
sns.rugplot(x='Word Count', data=sou)
sns.rugplot(y='Year', data=sou, color = sns.color_palette()[0])
plt.title('Speech Year Versus Word Count')
plt.xlabel('Word Count')
plt.ylabel('Year')
```

Commentary)

Early presidents from the late 1700s–1800s when the country was first founded often gave longer written messages, which explains the big cluster of high word counts in the older years.

In the mid-1900s and onward, speeches become shorter and more consistent, staying mostly below 10,000 words.

Recent speeches from like the 2000s–2020s cluster tightly around similar lengths, showing that modern presidents follow a more standard format.

#### Word Count Distribution per President

```python
# Hint: use pandas groupby to get mean word count per president then sort by order
order = sou.groupby('President')['Year'].max().sort_values(ascending=False).index
avg_counts = sou.groupby('President')['Word Count'].mean().reindex(order)
plt.figure(figsize=(15,5))
avg_counts.plot(kind='bar')
plt.title('Average State of the Union Word Count per President')
plt.xlabel('President')
plt.ylabel('Average Word Count')
plt.xticks(rotation=90)
plt.show()
```

Commentary)

Some presidents have unusually short average speeches (like early presidents who sometimes submitted very brief written messages), which makes them clear low outliers on the chart.

There’s a big jump in average length for a few specific presidents suggesting that probably some aspect of personal speaking/writing style plays a role, not just the era.

The bar heights move up and down a lot from president to president, showing that speech length isn’t really tied to political party or time period in a strict pattern.
